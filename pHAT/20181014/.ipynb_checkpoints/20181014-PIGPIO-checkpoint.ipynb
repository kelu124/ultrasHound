```python
#!/usr/bin/python
import spidev
#import RPi.GPIO as GPIO
import time

import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
FILE = "code/18khz.tmp"
with open(FILE) as f:
    content = f.readlines()
# you may also want to remove whitespace characters like `\n` at the end of each line
content = [x.strip() for x in content] 
```

```python
content[-1]
```

```python
A = []
for item in content:
    if len(item) < 7:
        resp = item.split(" ")
        A.append( 256*(int(resp[0])/8) + int(resp[1]) )
```

```python
plt.plot(A)
plt.show()
```

```python
F = np.abs(np.fft.fft(A))
```

```python
plt.plot(F[100:-100])
plt.show()
```

```python
plt.plot(F[1:len(F)/2-1])
plt.show()
```

```python
ax_idx = 3000+np.argmax(F[3000:100000])
print ax_idx
```

```python
T = len(F)
scale = float(18000.0/ax_idx)
print scale
```

```python
t = [ k*scale   for k in range(T)]
```

```python
plt.figure(figsize=(15,5))
plt.title("loglog plot of the "+FILE.split(".")[0])
plt.loglog(t[0:len(F)/2-1],F[0:len(F)/2-1])
plt.xlabel("Frequency (Hz)")
plt.savefig(FILE.split(".")[0]+'_loglog.png')
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(t[3000:30000],F[3000:30000])
plt.title("Peak for the "+FILE.split(".")[0])
plt.xlabel("Frequency (Hz)")
plt.savefig(FILE.split(".")[0]+'_detail.png')
plt.show()
```
