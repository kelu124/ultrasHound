```python
#!/usr/bin/python
import spidev
#import RPi.GPIO as GPIO
import time

import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open("15khz.tmp") as f:
    content = f.readlines()
# you may also want to remove whitespace characters like `\n` at the end of each line
content = [x.strip() for x in content] 
```

```python
content[-1]
```

```python
A = []
for item in content:
    if len(item) < 7:
        resp = item.split(" ")
        A.append( 256*(int(resp[0])/8) + int(resp[1]) )
```

```python
plt.plot(A)
plt.show()
```

```python
F = np.abs(np.fft.fft(A))
```

```python
plt.plot(F[1:-1])
plt.show()
```

```python
plt.plot(F[1:len(F)/2-1])
plt.show()
```

```python
ax_idx = np.argmax(F[10000:100000])
print ax_idx
```

```python
T = len(F)
scale = float(15000.0/ax_idx)
print scale
```

```python
t = [ k*scale   for k in range(T)]
```

```python
plt.plot(t[100:len(F)/2-1],F[100:len(F)/2-1])
plt.show()
```
