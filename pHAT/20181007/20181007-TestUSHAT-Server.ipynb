```python
#!/usr/bin/python
import spidev
import RPi.GPIO as GPIO
import time

import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
GPIO.setmode(GPIO.BCM)
ADC_CS = 22
LED1 = 17
LED2 = 27

GPIO.setup(LED1,GPIO.OUT)
GPIO.setup(LED2,GPIO.OUT)
GPIO.setup(ADC_CS,GPIO.IN, pull_up_down=GPIO.PUD_DOWN)
```

```python
print "Reset 25 - Low 1s"
GPIO.output(LED1,GPIO.LOW)
GPIO.output(LED2,GPIO.HIGH)
time.sleep(0.5)
print "Reset 25 - High 0.2s"
GPIO.output(LED2,GPIO.LOW)
GPIO.output(LED1,GPIO.HIGH)
time.sleep(0.5)
```

```python
spi = spidev.SpiDev()
spi.open(0,0) # CS2 - FPGA, on CE1 = IO4
spi.mode = 0b01
print "spi.cshigh is " + str(spi.cshigh)
print "spi mode is " + str(spi.mode)
```

```python
R = []
```

```python
spi.max_speed_hz
```

```python
spi.max_speed_hz = 181000
```

```python
A = []
for i in range(1000):
    to_send = [0x00,0x00]
    resp = spi.xfer2(to_send)
    A.append( 256*(resp[0]/8) + resp[1] )
```

```python
plt.plot(A)
plt.show()
```

```python
F = np.abs(np.fft.fft(A))
```

```python
plt.plot(F[1:-1])
plt.show()
```

```python
R.append([spi.max_speed_hz,A,F])
```

```python
np.save("sounds.npz", R)
```

```python
import json
```

```python
DATA = {}
for i in range(len(R)):
    #print R[i][0],R[i][1]
    DATA[R[i][0]] = {}
    DATA[R[i][0]]["A"] = np.asarray( R[i][1] ).tolist()
    DATA[R[i][0]]["F"] = np.asarray( R[i][2] ).tolist()
```

```python
with open('data.json', 'w') as outfile:
    json.dump(DATA, outfile)
```
