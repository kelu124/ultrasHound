```python
import time
import json
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
with open('data.json') as json_data:
    d = json.load(json_data)
print d.keys()
```

```python
for f in d.keys():
    t = d[f]["A"]
    FFT = np.abs(np.fft.fft(t))
    fft_shannnon = len(FFT)/2
    FIG, (ax1, ax2) = plt.subplots(1, 2,figsize=(15,5))
    ax1.plot(FFT[1:fft_shannnon])
    peak = np.argmax(FFT[1:fft_shannnon]) + 1
    ax1.set_title("SPI clk @"+f)
    ax2.plot(FFT[1:fft_shannnon])
    ax2.set_title("Peak @"+str(peak) + " - Sampling: "+str(int(f)/16))
    plt.show()
```

```python
for f in d.keys():

        t = d[f]["A"]
        signal = []
        for k in t:
            signal.append(k-255)
        FFT = np.abs(np.fft.fft(signal))
        peak = np.argmax(FFT[1:fft_shannnon/2]) + 1
            
        FMAX = (int(len(t)*10.0/peak)*100)/1000.0
        freq = range(len(t))
        x = [FMAX*i/len(t) for i in freq ]
        T = [i/FMAX for i in freq ]
        fft_shannnon = len(FFT)/2
        FIG, (ax1, ax2) = plt.subplots(1, 2,figsize=(15,5))
        ax1.plot(T,signal)

        ax1.set_title("Signal ( t in ms)")
        ax2.plot(x[1:fft_shannnon],FFT[1:fft_shannnon])
        
        TITLE2 = "Peak @"+str(peak) + " - Sampling: "+str(FMAX)+"khz (theory: "+str(int(f)/16)+")"
        TITLE2 = TITLE2 + "\nSPI clk @"+f
        
        ax2.set_title(TITLE2)
        plt.savefig("images/"+str(f)+".jpg", bbox_inches='tight')
        plt.show()
```
