```python
from scipy.io.wavfile import read as wavread
import matplotlib.pyplot as plt
import numpy as np

from scipy.signal import hilbert 
from scipy.signal import argrelextrema
```

```python
audiofilename = "Voix4-automatkolnhbf"
[samplerate, x] = wavread(audiofilename) # x is a numpy array of integer, representing the samples 
# scale to -1.0 -- 1.0
if x.dtype == 'int16':
    nb_bits = 16 # -> 16-bit wav files
elif x.dtype == 'int32':
    nb_bits = 32 # -> 32-bit wav files
max_nb_bit = float(2 ** (nb_bits - 1))
samples = x / (max_nb_bit + 1.0) # samples is a numpy array of float representing the samples 
print samplerate
oldx = x 
```

```python

print np.shape(samples)
x = []
for k in range(len(samples)):
    x.append(samples[k][1]+samples[k][0])
```

```python
t = range(len(x))
for k in range(len(x)):
    t[k] = 1.0*t[k] / samplerate
plt.plot(t,x)  # let's plot the sound
plt.xlabel("Time in s")
plt.show()
```

```python
L = len(samples)
F1 = 20600
F2 = 21000

f1 = int(F1*L/samplerate)
f2 = int(F2*L/samplerate)

f = range(len(samples))
for k in range(len(samples)):
    f[k] = 1.0e-3*f[k]*samplerate/len(samples)
```

```python
A = np.fft.fft(x)
plt.semilogy(f[f1:f2],np.abs(A)[f1:f2])  
plt.xlabel("Freq in kHz")

plt.savefig(audiofilename+'-1.png')
plt.show()
```

```python
B = np.fft.fft(x)
for i in range(L/2+1):
    if i < f1 or i > f2:
        B[i] = 0
        B[-i] = 0
```

```python
F = np.fft.ifft(B)
```

```python
plt.plot(t,x,"b")  # let's plot the sound
plt.plot(t,5.0*F,"r")  # let's plot the sound
plt.xlabel("Time in s")
plt.show()
```

```python
plt.plot(t[5900:76000],x[5900:76000],"b")  # let's plot the sound
plt.plot(t[5900:76000],5.0*F[5900:76000],"r")  # let's plot the sound
plt.xlabel("Time in s")
plt.show()
```

```python
L = len(x)
T = 8*1024
nS = int(L/T)
dd1 = int(F1*T/samplerate)
dd2 = int(F2*T/samplerate)

SP = []
RawSP = []
for i in range(nS):
    B = np.fft.fft(x[i*T:(i+1)*T-1])

    B = B / (np.average(np.abs(B[dd1:dd1+60])) + np.average(np.abs(B[dd1+100:dd1+160])) )
    SP.append(np.abs(B[dd1:dd2]))  
    RawSP.append( np.concatenate((B[dd1:dd2], B[T-dd2:T-dd1]), axis=0))
    
SPP = np.asarray(SP)

```

```python
plt.figure(figsize=(15,5))
plt.imshow(np.transpose(SPP), cmap='hot',vmin=SPP.min(),vmax=SPP.max()) 
plt.title("Plot of the "+str(F1/1000)+"kHz to "+str(F2/1000)+"kHz ("+audiofilename+")")
plt.savefig(audiofilename+'-2.png')
plt.show()
```
